## Циклы
***
Давайте посмотрим на самый простой цикл, который называется `for`:

```python
for i in range(5):
    print("Привет!")
```

Запустив ячейку сверху вы должны были увидеть, как питон 5 раз подряд печатает строку "Привет!".  
И вы наверняка догадались, что аргумент `5` в функции `range()` имеет к этому какое-то отношение.  
Так и есть: с каждым запуском цикла (с каждой печатью "Привет!") `i` — наша _переменная_ — увеличивается на 1, проходя все значения от 0 до 4 (так работает функция `range()` — она не включает последнее значение, то есть `range(5)` означает _"число от 0 до 5 не включая 5 само 5"_).

Маленькое задание: модифицируйте код из предыдущей ячейки, чтобы вместо "Привет!" на каждом шаге цикла программа выводила значение переменной `i`. Помните, что теперь вы хотите печатать не строку, а саму переменную, поэтому кавычки вам не нужны.

Переменная `i` служит _счетчиком_ для нашего цикла. Кроме этого, ее можно использовать для каких-то дополнительных целей внутри самого цикла.  
Давайте попробуем использовать цикл `for` для того, чтобы двигать Боба на заданное количество шагов в нужном направлении.  
Для этого в ячейке ниже введите следующий код:

***

&nbsp;&nbsp;&nbsp;&nbsp;Bob.move('Right')  
ThisLand.show_map()
***
*(Кстати, вы наверняка обратите внимание, что ноутбук сам сделат для вас отступ на следующей строке, когда вы напечатаете `for i in range(5):` и нажмете **Enter**. Это неспроста: отступы в Python используются для отделения блоков кода друг от друга. В нашем случае отступ дает понять питону, что `Bob.move('Right')` — это все еще тело цикла, а `ThisLand.show_map()` идет после того, как цикл закончился)*
***

Если все сделано правильно, то Боб будет перемещаться в заданном направлении на заданное в функции `range()` число шагов.

Цикл `for` является достаточно гибким для решения большинства типовых задач в программировании. Позднее вы узнаете, что `range()` это не единственный способ заставить цикл работать, но кажется, что это точно самый простой из них.

Теперь, когда мы научились создавать циклы, настала очередь упростить нашу жизнь до предела: в конце концов, если мы захотим подвигать Боба на 5 шагов вверх и на 7 вправо, то придется перепечатать цикл `for` два раза целиком, а перепечатывание — это именно то, от чего мы хотели отказаться, когда вводили циклы.  
Поскольку все, что мы хотим, это иметь возможность менять направление и число шагов, наверняка можно "завернуть" повторяющуюся часть кода (`for`, `range` и прочее) в какую-то обертку. Эти обертки и есть функции!

Давайте создадим свою собственную функцию. Это очень просто, выглядит похожим на цикл.  
Для объявления новой функции нам нужен новый _оператор_: `def`, а синтаксис выглядит так:

    def name_of_your_function(list, of, parameters):  
        what your function actually does

Как обычно, после двоеточия обязательно нужен отступ!  
Посмотрим на пример функции:

```python
def foo(n):
    for i in range(n):
        print('Foo')
```

Как вы думаете, что эта функция делает?  
Проще всего не гадать, а посмотреть. А для этого функцию надо _вызвать_.  
Поскольку вы уже знаете, как это сделать, вызовите `foo` в следующей ячейке.  
Не забудьте указать аргумент! Поскольку `n` в теле функции передается `range()`, это должно быть число, а не строка.

Что здесь произошло? Мы завернули довольно громоздкую конструкцию с циклом `for` внутрь функции, так что теперь нам достаточно поменять `n` в вызове функции `foo`, а не перепечатывать весь цикл.

То же самое можно сделать и с Бобом! Давайте напишем функцию `move_steps_right(steps)`.  
Она будет принимать на входе количество шагов, класть это значение в переменную `steps`, а затем передавать ее в цикл.

Теперь вызовите вашу функцию несколько раз, чтобы проверить, что все работает, как задумано.

Надеемся, что у вас все получилось. Теперь мы вооружены функцией `move_steps_right()`, которая двигает Боба вправо на нужное количество шагов. Но зачем останавливаться на достигнутом? Ведь помимо количества шагов мы можем передавать в функцию и направление! Тогда не придется перепечатывать 4 почти одинаковые функции (`move_steps_right`, `move_steps_left`, и так далее).

Нет, гораздо лучше будет ввести функцию `move_steps()`, которая будет принимать сразу два аргумента:

## Условия
***
И вновь мы смотрим на новый синтаксис, который почему-то очень похож на старый (вспомните цикл `for`): те же блоки с отступами, те же двоеточия. Но все-таки есть и что-то такое, с чем мы еще не сталкивались.

    if condition:  
        outcome_if_condition_is_True 
    else:
        outcome_if_condition_is_False

Вот, как это может выглядеть в жизни (не забудьте запускать ячейки!)):

```python
if 2>1:
    print("Да-да, 2 действительно больше 1")
else:
    print("А у вас точно не двойка по математике?")
```

```python
if 3>5:
    print("Мне кажется, что математика работает как-то не так...")
else:
    print("Ну да, все верно, 3 действительно меньше 5")
```

Можно даже делать так:

```python
foo = True
if foo:
    print('bar')
else:
    print('baz')
```

Как видите, любая конструкция с `if` содержит _условие_ (condition) и два результата: в случае, если условие верно (`True`) и если неверно (`False`).

***

Если вы освоились с условным оператором, можно вернуться к Бобу. Как мы уже видили, функция `Bob.look()` возвращает значения `True` или `False` в зависимости от того, проходима ли клетка, на которую Боб смотрит. Это значит, что такую функцию можно использовать в качестве _условия_ в операторе `if`:

```python
if Bob.look('right'):
    Bob.move('right')
else:
    print("Я не могу там пройти. И что же мне теперь делать?..")
```

Как видите, когда Боб смотрит направо, он видит, что клетка там занята, поэтому наше _условие_ возвращается как `False`, поэтому Боб окунается в пучину неуверенности в себе.

Один цикл мы сегодня уже видели — это `for`. Но он не единственный в Python!  
Поприветствуйте цикл `while` — отличное решение для тех случаев, когда вы не знаете точно, сколько раз нужно повторить действие, но знаете, когда надо остановиться.

Выглядит этот цикл так:  

    while condition:
        code

И так он и работает: перед тем, как заново запустить _тело_ цикла, проверяется _условие_ (condition).  
С английского while переводится как "до тех пор, пока", то есть пока *условие* будет возвращать `True`, цикл будет повторяться и прервется только тогда, когда *условие* станет `False`.  

Например:

```python
i = 7
while i > 1:
    print("There were {} in the bed, and the little one said 'Roll over! Roll over!'\nThey all rolled over, and 1 fell out! La-la-la-la-la-la-la-la!".format(i))
    i = i-1 # Эта переменная будет становится меньше на 1 с каждым повторением тела цикла
```

Перед вами текст английской детской песенки про животных, которые падают с кровати, пока не останется только один.  
Как видите, `while` — идеальный цикл для этой задачи: как только количество животных стало меньше 1, цикл остановился.

Опасность `while` в том, что при неправильно заданном условии вы рискуете запустить бесконечный цикл. Остановить его можно только перезапустив ноутбук (Kernel → restart), поэтому будьте внимательны!
