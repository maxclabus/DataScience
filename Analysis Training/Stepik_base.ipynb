```python
concentrations.query('genus == "Fucus"').agg(['min','mean','max']).alanin.round(2)
```

```python
my_stat = pd.read_clipboard(index_col = 0)
```

```python
my_stat
```

```python
my_stat = my_stat.fillna(0)
```

```python
n_users_median = my_stat[my_stat.n_users >= 0].n_users.median()
```

```python
my_stat
```

```python
my_stat.loc[my_stat.n_users < 0, 'n_users'] = n_users_median
```

```python
my_stat
```

```python
mean_session_value_data = my_stat.groupby('group', as_index = False).agg({'session_value':'mean'}).rename(columns = {'session_value':'mean_session_value'})
```
