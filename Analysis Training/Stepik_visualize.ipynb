```python
Students = pd.read_csv('CsvFiles/StudentsPerformance.csv')
Students.columns = Students.columns.str.replace(' ', '_')
```

```python
Students.lunch.value_counts().plot.pie();
```

```python
Students.plot.scatter(x = 'math_score', y = 'reading_score');
```

```python
ax = sns.lmplot(x = 'math_score', y = 'writing_score', data = Students.query('math_score > 70'), hue = 'gender', aspect = 2);
ax.set_xlabels('Math Score');
ax.set_ylabels('Writing Score');
```
